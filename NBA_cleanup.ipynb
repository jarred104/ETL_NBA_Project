```python
import pandas as pd
import unidecode
```

```python
# File to Load (Remember to Change These)
file_to_load = "player_stats_2014.csv"

# Read Purchasing File and store into Pandas data frame
player_data = pd.read_csv(file_to_load)
```

```python
player_data.head()
```

```python
player_data['player_lastname'] = player_data['player_lastname'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
player_data['player_firstname'] = player_data['player_firstname'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```
