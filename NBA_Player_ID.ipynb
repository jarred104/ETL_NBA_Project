```python
import pandas as pd
```

```python
# File to Load
player_data_2014_csv = "player_stats_2014_normal.csv"
player_data_2015_csv = "player_stats_2015_normal.csv"
player_data_2016_csv = "player_stats_2016_normal.csv"
player_data_2017_csv = "player_stats_2017_normal.csv"
player_data_2018_csv = "player_stats_2018_normal.csv"
player_data_2019_csv = "player_stats_2019_normal.csv"

# Read Data to DataFrames
player_data_2014 = pd.read_csv(player_data_2014_csv)
player_data_2015 = pd.read_csv(player_data_2015_csv)
player_data_2016 = pd.read_csv(player_data_2016_csv)
player_data_2017 = pd.read_csv(player_data_2017_csv)
player_data_2018 = pd.read_csv(player_data_2018_csv)
player_data_2019 = pd.read_csv(player_data_2019_csv)
```

```python
# Check Player Data
player_data_2014
```

```python
# Concatenate All Years of Data 
all_df = [player_data_2014, player_data_2015, player_data_2016, player_data_2017, player_data_2018, player_data_2019]
all_player_data = pd.concat(all_df, sort=False)
```

```python
# Check All Data
all_player_data = all_player_data.sort_values('player_lastname')
all_player_data
```

```python
# Build DataFrame from ID, Last Name, and First Name
player_id_df = all_player_data[['player_id','player_lastname','player_firstname']]
```

```python
player_id_df.head()
```

```python
# Drop Duplicates
player_id_clean = player_id_df.drop_duplicates()
```

```python
player_id_clean.set_index('player_id')
```

```python
# Save to .csv
player_id_clean.to_csv(r'player_id.csv',index=False)
```

```python
draymond_csv = "draymond_score.csv"
draymond_df = pd.read_csv(draymond_csv)

draymond_df['player_lastname'] = draymond_df['player_lastname'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
draymond_df['player_firstname'] = draymond_df['player_firstname'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

draymond_df_sort = draymond_df.sort_values('player_lastname')
draymond_df_sort
```

```python
draymond_clean = pd.merge(player_id_clean, draymond_df_sort, how = 'right', on = ['player_lastname', 'player_firstname'])
```

```python
draymond_clean.sort_values('player_lastname')
```

```python
draymond_clean.to_csv(r'draymond_clean.csv',index=False)
```
